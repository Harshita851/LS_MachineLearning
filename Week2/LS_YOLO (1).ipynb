```python
import ultralytics
from ultralytics import YOLO
```

```python
data_yaml = '/content/drive/MyDrive/YOLOv8Dataset/car/data.yaml'
```

```python
model = YOLO('yolov8n.pt')
model.train(data=data_yaml, epochs=20, imgsz=640, batch=16, name='yolo_custom')
```

```python
results = model.val()
```
