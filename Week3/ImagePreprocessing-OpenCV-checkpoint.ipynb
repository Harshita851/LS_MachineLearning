## Code in the following 3 cells only

### Cartoon
For the bart image, you need to mask the dress. Write all your code in the cell below and the final image should be stored in ```cartoon_y_mentee``` variable only or it'll show error later.

````python
#START CODE HERE (Do all your processing here)

hsv_image = cv2.cvtColor(cartoon, cv2.COLOR_BGR2HSV)

lower_green = np.array([35, 100, 100])
upper_green = np.array([85, 255, 255])

mask = cv2.inRange(hsv_image, lower_green, upper_green)


green_regions = cv2.bitwise_and(cartoon, cartoon, mask=mask)

cartoon_rgb = cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB)

cartoon_y_mentee  = cv2.cvtColor(green_regions, cv2.COLOR_BGR2RGB)

# Display the results
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Original image
axs[0].imshow(cartoon_rgb)
axs[0].set_title("Original Image")
axs[0].axis("off")

# Mask
axs[1].imshow(mask, cmap='gray')
axs[1].set_title("Mask")
axs[1].axis("off")

# Green regions
axs[2].imshow(cartoon_y_mentee )
axs[2].set_title("Green Regions")
axs[2].axis("off")

plt.show()




#END CODE HERE (the final processed image should be cartoon_y_mentee)
````

### Girl
For the girl image, you need to detect the edges. Write all your code in the cell below and the final image should be stored in ```girl_y_mentee``` variable only or it'll show error later.

````python
#START CODE HERE (Do all your processing here)


gray_image = cv2.cvtColor(girl, cv2.COLOR_BGR2GRAY)


k = 5
blurred_image = cv2.medianBlur(gray_image, k)


girl_y_mentee = cv2.Canny(blurred_image, 100, 200)


fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(cv2.cvtColor(girl, cv2.COLOR_BGR2RGB))
axs[0].set_title("Original Image")
axs[0].axis("off")


axs[1].imshow(blurred_image, cmap='gray')
axs[1].set_title("Blurred Image")
axs[1].axis("off")


axs[2].imshow(girl_y_mentee, cmap='gray')
axs[2].set_title("Edge Detection")
axs[2].axis("off")

plt.show()





#END CODE HERE (the final processed image should be girl_y_mentee)
````

### Fingerprint
For the fingerprint image, you need to convert it to binary image. Binary images have 2 intensities only (0 or 255). Write all your code in the cell below and the final image should be stored in ```cartoon_y_mentee``` variable only or it'll show error later.

````python
#START CODE HERE (Do all your processing here)

_, binary_threshold = cv2.threshold(fingerprint, 110, 255, cv2.THRESH_BINARY)

fingerprint_y_mentee = cv2.bitwise_not(binary_threshold)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(fingerprint, cmap='gray')
axs[0].set_title("Original Image")
axs[0].axis("off")

# Binary Thresholding
axs[1].imshow(fingerprint_y_mentee, cmap='gray')
axs[1].set_title("Binary Thresholding")
axs[1].axis("off")

plt.show()

fingerprint_y_mentee = cv2.bitwise_not(binary_threshold) #(in Binary format only i.e. pixel intensity = 0 or 255)

#END CODE HERE (the final processed image should be fingerprint_y_mentee)
````
